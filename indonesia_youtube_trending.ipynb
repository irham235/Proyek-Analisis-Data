### Data Preprocessing

```python
import pandas as pd

# input data
data = pd.read_csv('/content/trending.csv')
data.head()
```

```python
# tampilkan kolom
data.columns
```

```python
# buat data baru berdasarkan kolom yang diinginkan saja

data = data[['title', 'channel_name', 'category_id', 'view', 'like',
       'dislike', 'comment', 'trending_time']]
data.head()
```

```python
# ubah category_id menjadi category_name
# tetapi sebelum itu harus meng import category_names terlebih dahulu

import json

with open("/content/category.json", "rb") as f:
  categories = json.load(f)
```

```python
categories
```

```python
categories = pd.json_normalize(categories, record_path="items")
categories_dict = {idx:title for idx, title in zip(categories['id'], categories['snippet.title'])}
```

```python
from re import X
# ubah data category_id menjadi category di kolom baru
data['category'] = [categories_dict[str(x)] if str(x) in list(categories_dict.keys()) else "Unknown" for x in data['category_id']]
```

```python
data
```

```python
# lihat semua kategori dan jumlah mereka
data['category'].value_counts()
```

```python
# cek jika ada nilai yang bernilai NA
data.isnull().sum()
```

```python
# mengubah semua nilai bernilai NA menjadi 0
data = data.fillna(0)
```

```python
# cek kembali
data.isnull().sum()
```

```python
# mencari nilai trending_date dan menghapus trending_time
trending = pd.to_datetime(data['trending_time'])
data['trending_date'] = trending.dt.date
data = data.drop(['trending_time'], axis=1)
```

```python
data
```

```python
# melihat daya duplikat
data[['title', 'channel_name']].value_counts().reset_index(name='counts')
```

```python
# hapus semua data duplikat
used_data = data.drop_duplicates(['title', 'channel_name'], keep='first')
```

```python
used_data = used_data.reset_index(drop=True)
```

```python
used_data
```

```python
# menghapus category_id karena tidak dibutuhkan

used_data = used_data.drop(['category_id'], axis=1)
used_data
```

menyimpan semua file ke bentuk pickle

mengapa menggunakan pickle, karena data dengan format pickle tidak mengubah bentuk dari data tersebut, pickle juga memiliki ukuran yang sangat kecil, jadi bisa menyimpan data yang besar di memori yang lebih kecil

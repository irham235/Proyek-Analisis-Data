```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(15, 5, 250)

plt.hist(x=x, bins=15)
plt.show()
```

Menggunakan library seaborn

```python
import seaborn as sns

sns.histplot(x=x, bins=15)
plt.show()
```

mempermudah dalam mengidentifikasi distribusi dari sebuah data kuantitatif, menambahkan parameter "kde=True"

```python
sns.histplot(x=x, bins=15, kde=True)
plt.show()
```
